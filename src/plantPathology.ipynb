```python
import cv2.cv2 as cv2
import pandas as pd
import matplotlib.pyplot as plt


def showImages(images: list):
    rows = len(images)
    for num, img in enumerate(images):
        f = plt.imread('./data/images/{0}.jpg'.format(img))
        plt.subplot(rows / 4 + 1, 4, num + 1)
        plt.tight_layout(pad=1)
        plt.title(img.split('.')[0])
        plt.axis('off')
        plt.imshow(f)

data = pd.read_csv('./data/train.csv')

data = data[['image_id','healthy','multiple_diseases']]
#showImages(data[:20]['image_id'])
```
